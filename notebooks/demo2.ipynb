```python
import pandas
```

```python
# data = pandas.read_csv("\\Users\\shidiq/Library/CloudStorage/OneDrive-Personal/Personal Projects/Lecture/pengenalan python untuk data science/data/data-time-series.csv")
data = pandas.read_csv("../data/data-time-series.csv")
data.head(5)
```

```python
x = data["range"].values
x
```

```python
y = data["score"].values
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, c="#FF6B6B", marker="^", s=90)
```

```python
from scipy.optimize import curve_fit
import numpy
```

```python
def objective(x, a, b, c, d, e, f):
    return (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f


popt, _ = curve_fit(objective, x, y)
a, b, c, d, e, f = popt

x_line = numpy.arange(20, 460, 1)
y_line = objective(x_line, a, b, c, d, e, f)
```

```python
plt.plot(x_line, y_line)
```

```python
fig, ax = plt.subplots(figsize=[10.72, 8.205])
ax.scatter(x, y, c="#FF6B6B", marker="^", s=90, label="Score")
ax.plot(x_line, y_line, label="Fitting")
ax.legend(loc=4)
ax.set_xlabel("Range")
ax.set_ylabel("Score")
```
